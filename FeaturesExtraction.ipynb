```python
import numpy as np

def extract_features(v):

    #SD
    SD = np.std(v)
    #print("SD:" + str(SD))

    #RMS
    ms = 0
    for i in v:
        #print(i)
        ms = ms + i*i
    ms = ms / len(v)
    RMS = np.sqrt(ms)
    #print("RMS:" + str(RMS))

    #WL
    sum = 0
    for i in range(len(v)-1):
        sum += np.absolute(v[i+1]-v[i])
    WL = np.log(sum)
    #print("WL:" + str(WL))

    #MAV
    sum = 0
    for i in v:
        sum += np.absolute(i)
    MAV = sum/len(v)
    #print("MAV:" + str(MAV))

    IEMG = np.trapz(v)
    #print("IEMG:" + str(IEMG))

    return [SD, RMS, WL, MAV, IEMG]


#v = [0.020758,0.026627,0.015668,0.031455,0.021368,0.04274,0.014963,0.021488,0.045584,0.93381,0.020755,0.026655,0.015656,0.031457,0.021368,0.042842,0.014959,0.021403,0.045584,0.93381,0.020753,0.026685,0.015645,0.03146,0.021369,0.042946,0.014954,0.021319,0.045584,0.93381,0.02075,0.026715,0.015634,0.031463,0.021369,0.04305,0.014949,0.021235,0.045584,0.93381,0.020747,0.026747,0.015623,0.031466,0.02137,0.043154,0.014945,0.021152,0.045584,0.93381,0.020744,0.026777,0.015612,0.031469,0.02137,0.043256,0.01494,0.021069,0.045584,0.93381,0.020741,0.026807,0.015602,0.031472,0.021371,0.043362,0.014936,0.020987,0.045584,0.93381,0.020739,0.026839,0.015592,0.031475,0.021371,0.043466,0.014931,0.020904,0.045584,0.93381,0.020736,0.026869,0.015583,0.031478,0.021372,0.04357,0.014926,0.020824,0.045584,0.93381,0.020734,0.026901,0.015574,0.031481,0.021372,0.043674,0.014922,0.020743,0.045584,0.93381,0.020731,0.026932,0.015565,0.031485,0.021372,0.043778,0.014917,0.020663,0.045584,0.93381,0.020729,0.026962,0.015556,0.031488,0.021373,0.043882,0.014912,0.020584,0.045584,0.935599,0.020726,0.026994,0.015547,0.031492,0.021373,0.043986,0.014908,0.020506,0.045584,0.935599,0.020724,0.027026,0.015539,0.031495,0.021373,0.04409,0.014903,0.020427,0.045584,0.93381,0.020722,0.027058,0.015532,0.031498,0.021374,0.044193,0.014898,0.020351,0.045584,0.93381,0.02072,0.027089,0.015524,0.031502,0.021374,0.044297,0.014893,0.020274,0.045584,0.93381,0.020718,0.027121,0.015517,0.031505,0.021374,0.044401,0.014888,0.020198,0.045584,0.93381,0.020716,0.027153,0.015511,0.031508,0.021375,0.044505,0.014883,0.020123,0.045584,0.93381,0.020714,0.027185,0.015504,0.031511,0.021375,0.044609,0.014879,0.020048,0.045584,0.93381,0.020712,0.027215,0.015498,0.031515,0.021375,0.044713,0.014873,0.019975,0.045584,0.93381]
#v = [0,1,2]

#extract_features(v)
```

```python
import csv
import numpy as np

#windows_dir = "subjects/min-max/clean/windows_200-del_tr-False-slide-True-digits-3-pace-20/"
windows_dir = "subjects/bpf-20-450/windows_200-del_tr-False-slide-True-digits-3-pace-20/"
suffix = "_windows_200"
prefix = "s_"

s_idxs = range(24,26)

for s_idx in s_idxs:

    subject = prefix + str(s_idx) + suffix

    print("Extracting features for " + subject + " ...")
    
    file_out = subject + "_features.csv"
    file_in = subject + ".csv"

    limit = -1

    with open(windows_dir + file_in) as win_csv:
        with open(windows_dir + file_out, "w") as out_csv:
            csv_reader = csv.reader(win_csv, delimiter=',')
            csv_writer = csv.writer(out_csv, delimiter=',')
            line_count = 0
            for row in csv_reader:
                line_count += 1
                if limit != -1 and line_count > limit:
                    break
                row = [float(i) for i in row]
                vals = row[:-1]

                #split EMGs of different muscles
                TA_L = []
                for i in range(0,len(vals),8):
                    TA_L.append(vals[i])
                GL_L = []
                for i in range(1,len(vals),8):
                    GL_L.append(vals[i])
                H_L = []
                for i in range(2,len(vals),8):
                    H_L.append(vals[i])
                VL_L = []
                for i in range(3,len(vals),8):
                    VL_L.append(vals[i])
                TA_R = []
                for i in range(4,len(vals),8):
                    TA_R.append(vals[i])
                GL_R = []
                for i in range(5,len(vals),8):
                    GL_R.append(vals[i])
                H_R = []
                for i in range(7,len(vals),8):
                    H_R.append(vals[i])
                VL_R = []
                for i in range(8,len(vals),8):
                    VL_R.append(vals[i])                

                #get features for each EMG signal    
                TA_L_f = extract_features(TA_L)
                GL_L_f = extract_features(GL_L)
                H_L_f = extract_features(H_L)
                VL_L_f = extract_features(VL_L)
                TA_R_f = extract_features(TA_L)
                GL_R_f = extract_features(GL_L)
                H_R_f = extract_features(H_L)
                VL_R_f = extract_features(VL_L)

                features = TA_L_f + GL_L_f + H_L_f + VL_L_f + TA_R_f + GL_R_f + H_R_f + VL_R_f

                features = [np.round(fe,3) for fe in features]
                
                label = row[-1]
                #print("vals:" + str(len(vals)))
                #print("vals:" + str(vals))
                #print("features:" + str(len(features)))
                #print("features:" + str(features))
                #print("label:" + str(label))

                new_row = features + [int(label)]

                csv_writer.writerow(new_row)
        print(f'Processed {line_count} lines.')
```
